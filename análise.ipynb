```python
import pandas as pd
from google.colab import files
import matplotlib.pyplot as plt

dados_total_2020 = {}

uploaded = files.upload()
nome_arquivo = list(uploaded.keys())[0]

dados_total_2020[f'planilha_{i}'] = pd.read_csv(nome_arquivo, sep=';')

dados_combinados_2020 = pd.concat(dados_total_2020.values())

soma_total_mortes = dados_combinados_2020['ano'].sum()

print("no ano de 2020 morreram :", soma_total_mortes," pessoass")

dados_combinados_2020 = dados_combinados_2020.sort_values(by='ano_percentual', ascending=False)

labels = dados_combinados_2020['filtros.indicador.nome']
sizes = dados_combinados_2020['ano_percentual']

plt.figure(figsize=(10, 6))
plt.barh(labels, sizes, color='skyblue')

plt.title('Porcentagem de Mortes por Cap√≠tulo do CID-10 no ano de 2020 para todas faixas et√°rias')
plt.xlabel('Porcentagem (%)')
plt.ylabel('Cap√≠tulo do CID-10')

for index, value in enumerate(sizes):
    plt.text(value, index, f'{value:.2f}%', color='black', ha='left', va='center')

plt.tight_layout()
plt.show()
```
